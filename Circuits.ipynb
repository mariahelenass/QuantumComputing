```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram, plot_distribution
from qiskit import transpile
```

```python
# simulador
simulator = BasicSimulator()
```

```python
circuit = QuantumCircuit(2,2) # 2 qubits e 2 bits
circuit.h(0) # hadamard
circuit.measure(0,0)
circuit.measure(1,1)
circuit.draw("mpl")
```

Uma linha - bit quântico

Duas linhas - bit clássico

```python
# rodando o circuito 10 vezes
job = simulator.run(circuit, shots=10, memory=True)
result = job.result()
```

Quanto maior o número de shots, melhor

```python
count = result.get_counts()
print(count)
```

### No qubit 0:

Foi medido 4 vezes no 1 e 6 vezes no 0.

### No qubit 1:

Apenas zero.


```python
# "quase probabilidade"
plot_distribution(count)
```

```python
qc = QuantumCircuit(2,2)
# rotacao no qubit 0
qc.rx(np.pi/3, 0)
qc.measure(range(2), range(2))
qc.draw("mpl")
```

```python
t = transpile(qc, backend=simulator)
t.draw("mpl")
```

```python
job = simulator.run(circuit, shots=10, memory=True)
result = job.result()
```
